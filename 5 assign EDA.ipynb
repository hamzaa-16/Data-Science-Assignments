```python
# Calculating IQR
Q1 = df['LB'].quantile(0.25)
Q3 = df['LB'].quantile(0.75)
IQR = Q3 - Q1

# Identifying outliers
outliers = df[(df['LB'] < (Q1 - 1.5 * IQR)) | (df['LB'] > (Q3 + 1.5 * IQR))]
print(outliers)
```

```python
# Replace outliers with median
median = df['LB'].median()
df['LB'] = np.where((df['LB'] < (Q1 - 1.5 * IQR)) | (df['LB'] > (Q3 + 1.5 * IQR)), median, df['LB'])
print(df)
```

### line plots for identifying trends


```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named 'df'
def plot_all_line_plots(df):
    numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns
    
    for column in numerical_columns:
        plt.figure()
        plt.plot(df[column], marker='o')
        plt.title(f'Line Plot of {column}')
        plt.xlabel('Index')
        plt.ylabel(column)
        plt.grid(True)
        plt.show()

# Example usage:
plot_all_line_plots(df)
```
